# Cleaning

```python
citibike.head()
```

```python
#checking nulls again
citibike.isnull().sum()
```

```python
#making a subset of nan values in year of. birth
nan_birth = citibike[citibike['birth_year'].isnull() ==True]
```

```python
nan_birth.shape
```

```python
#dropping Nan Values
citi_clean = citibike[citibike['birth_year'].isnull()== False]
```

```python
#checking on successful drop
citi_clean.shape
```

```python
citi_clean.head()
```

```python
citi_clean = citi_clean.drop(columns = ['bike_id'])
```

```python
citi_clean = citi_clean.drop(columns = ['trip_id'])
```

```python
citi_clean.info()
```

```python
#making a subset of users born after 1922
citi_clean_1 = citi_clean.loc[citi_clean['birth_year'] > 1922]
```

```python
citi_clean_1.head()
```

```python
# making sure there are 7 days of the week
citi_clean_1['day_of_the_week'].unique()
```

```python
#changing gender type to string
citi_clean_1['gender'] = citi_clean_1['gender'].astype('str')
```

```python
#changing gender variables to the name of gender
citi_clean_1['gender']=citi_clean_1['gender'].replace([0,1,2],['Uknown','male','female'])
```

```python
citi_clean_1.describe()
```

```python
citi_clean_1.head()
```

```python
#changing start time to start of ride and making the datatype - datetime
citi_clean_1['start_of_ride']=pd.to_datetime(citi_clean_1['start_time'],infer_datetime_format=True) 
```

```python
#checking code
citi_clean_1.head()
```

```python
#checking data type
citi_clean_1.info()
```

```python
#changing end time to end of ride and making the datatype - datetime
citi_clean_1['end_of_ride']=pd.to_datetime(citi_clean_1['end_time'],infer_datetime_format=True) 
```

```python
citi_clean_1.head()
```

```python
#checking data types
citi_clean_1.info()
```

```python
#dropping start time column
citi_clean_1 = citi_clean_1.drop(columns = ['start_time'])
```

```python
#dropping end time column
citi_clean_1 = citi_clean_1.drop(columns = ['end_time'])
```
